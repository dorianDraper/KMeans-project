# Week 3 - Create a Labelled Plot of the Happiness Data

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv('../dataset.csv')
data.head()
```

```python
# We want to analyze the data, so we will use the describe() method
data.describe()
```

```python
data.shape
```

```python
data.info()
```

```python
happy = data['happyScore']
income = data['avg_income']
```

```python
# lets plot the data
plt.scatter(income, happy)
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.title('Income vs Happiness Score')
plt.show()
```

```python
# Sort data by income, inplace = True to modify the original data
data.sort_values('avg_income', inplace=True)
data.head()
```

```python
data.reset_index(drop=True, inplace=True) # reset the index after sorting because it is not in order anymore
data.head()
```

```python
# Now our data is sorted by income, we can see the relationship between income and happiness score
plt.scatter(data['avg_income'], data['happyScore'])
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.title('Income vs Happiness Score')
plt.show()
```

```python
# We can see that there is a positive correlation between income and happiness score
# Now we will use the KMeans algorithm to cluster the data
from sklearn.cluster import KMeans

# We will use the elbow method to find the optimal number of clusters
wcss = [] # Within-Cluster-Sum-of-Squares (WCSS)
for i in range(1, 11):  # We will try from 1 to 10 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)  # k-means++ is an optimized way to initialize the centroids, random_state is the seed for the random number generator
    kmeans.fit(data[['avg_income', 'happyScore']]) # We only use the income and happiness score columns to cluster the data
    wcss.append(kmeans.inertia_)  # inertia_ is the WCSS value

# Plot the elbow method
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# We can see that the optimal number of clusters is 3 because the WCSS value does not decrease significantly after 3 clusters 
# Now we will fit the KMeans algorithm with 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42) # We will use 3 clusters 
data['cluster'] = kmeans.fit_predict(data[['avg_income', 'happyScore']])  # We add a new column 'cluster' to the data with the cluster number for each data point
data.head()
```

```python
# Now we will plot the data with the clusters
plt.scatter(data[data['cluster'] == 0]['avg_income'], data[data['cluster'] == 0]['happyScore'], color='red', label='Cluster 1')
plt.scatter(data[data['cluster'] == 1]['avg_income'], data[data['cluster'] == 1]['happyScore'], color='blue', label='Cluster 2')
plt.scatter(data[data['cluster'] == 2]['avg_income'], data[data['cluster'] == 2]['happyScore'], color='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='yellow', label='Centroids', s=100)
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.title('Income vs Happiness Score')
plt.legend()
plt.show()
```

```python
# We want to add in the plot labels for the highest income and happiness score countries for each cluster 
# We will use the annotate() function to add text to the plot
# We will find the highest income and happiness score countries for each cluster
cluster_0 = data[data['cluster'] == 0].sort_values('avg_income', ascending=False).head(1)
cluster_1 = data[data['cluster'] == 1].sort_values('avg_income', ascending=False).head(1)
cluster_2 = data[data['cluster'] == 2].sort_values('avg_income', ascending=False).head(1)

# Plot the data with the clusters and labels
plt.scatter(data[data['cluster'] == 0]['avg_income'], data[data['cluster'] == 0]['happyScore'], color='red', label='Cluster 1')
plt.scatter(data[data['cluster'] == 1]['avg_income'], data[data['cluster'] == 1]['happyScore'], color='blue', label='Cluster 2')
plt.scatter(data[data['cluster'] == 2]['avg_income'], data[data['cluster'] == 2]['happyScore'], color='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='orange', label='Centroids', s=100)
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.title('Income vs Happiness Score')
plt.annotate(f'{cluster_0.iloc[0]["country"]}', (cluster_0.iloc[0]['avg_income'], cluster_0.iloc[0]['happyScore']), color='red')
plt.annotate(f'{cluster_1.iloc[0]["country"]}', (cluster_1.iloc[0]['avg_income'], cluster_1.iloc[0]['happyScore']), color='blue')
plt.annotate(f'{cluster_2.iloc[0]["country"]}', (cluster_2.iloc[0]['avg_income'], cluster_2.iloc[0]['happyScore']), color='green')
plt.legend()
plt.show()
```
