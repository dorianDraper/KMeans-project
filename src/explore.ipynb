# Explore KMeans

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generate sample data
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Fit the KMeans model to the data
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)
```

```python
# Making predictions with new data
new_data = np.array([[0, 0], [4, 4]])
predictions = model.predict(new_data)
```

```python
# Print the predictions
print(predictions)
```

```python
# Plotting the data
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=model.labels_)
plt.scatter(new_data[:, 0], new_data[:, 1], c='r', marker='x')
plt.show()
```

<p style="color: pink">The KMeans model has clustered the data into three clusters. The new data points are assigned to the clusters based on the closest centroid. The red crosses represent the new data points.</p>
